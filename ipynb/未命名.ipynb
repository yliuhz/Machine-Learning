```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
x = np.linspace(-10,10)
y = 1/(1+np.exp(-x))
plt.plot(x,y)
plt.title('sigmoid')
os.makedirs('./pics', exist_ok=True)
plt.savefig('./pics/sigmoid.png')
```

```python
x = np.linspace(-10,10)
y = np.log(1+np.exp(x))
plt.plot(x,y)
plt.title('softplus')
plt.savefig('./pics/softplus.png')
```

```python
x = np.linspace(0.1,0.9)
y = np.log(x/(1-x))
z = 1/(1+np.exp(-x))
plt.plot(x,y,'r',label='logit')
plt.plot(x,z,'b', label='sigmoid')
plt.plot(x,x,'g--', label='y=x')
plt.xlim(0,4)
plt.legend()
plt.savefig('pics/logit.png')
```

```python
def sigma(x):
    return 1/(1+np.exp(-x))
def r_sigma(x):
    return np.log(x/(1-x))
```

```python
sigma(r_sigma(0.9))
```

```python
r_sigma(sigma(0.9))
```
