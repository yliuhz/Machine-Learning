```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
x = np.linspace(-10,10)
y = 1/(1+np.exp(-x))
plt.plot(x,y)
plt.title('sigmoid')
os.makedirs('./pics', exist_ok=True)
plt.savefig('./pics/sigmoid.png')
```

```python
x = np.linspace(-10,10)
y = np.log(1+np.exp(x))
plt.plot(x,y)
plt.title('softplus')
plt.savefig('./pics/softplus.png')
```
